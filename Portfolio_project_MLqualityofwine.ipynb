# Scaling our variables

Scaling training data

```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt 

scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_train
# Scaling help stablishing a better range for the difference between values in the data
```

Scaling test data

```python
scaled_X_test = scaler.transform(X_test)
scaled_X_test
```

Importing the ML algorithm

```python
from sklearn.svm import SVC #support vector classifier

svc_clf = SVC()
svc_clf.fit(scaled_X_train, y_train) #take the scaled x training data and fit it using the y training data

```

```python
svc_clf_predictions = svc_clf.predict(scaled_X_test) #Take the test scaled data and use the previous fit to predict how y should be
svc_clf_predictions
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
print(confusion_matrix(y_test,svc_clf_predictions))
plt.matshow(confusion_matrix(y_test,svc_clf_predictions))
```

```python
print(accuracy_score(y_test, svc_clf_predictions))
print(precision_score(y_test, svc_clf_predictions))
print(recall_score(y_test, svc_clf_predictions))
```

```python
def algo_accuracy(y_test, predictions): #This wil create a function that compares the test and the predicted value of y 
    conf_matrix = confusion_matrix(y_test, predictions)
    accuracy = accuracy_score(y_test, predictions)
    precision = precision_score(y_test, predictions)
    recall = recall_score(y_test, predictions)

    return confusion_matrix, accuracy, precision, recall
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

param_grid =[
    {'n_estimators':  [3, 10, 30]}, 
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},]

rfc = RandomForestClassifier()
grid_search = GridSearchCV(rfc, param_grid, cv=5,
                            scoring = 'accuracy',
                            return_train_score=True)
grid_search.fit(scaled_X_train, y_train)    
```

```python
rfc_clf = grid_search.best_estimator_
rfc_clf
```

```python
rfc_predictions = rfc_clf.predict(scaled_X_test)
rfc_predictions
```

```python
algo_accuracy(y_test, rfc_predictions)
```
